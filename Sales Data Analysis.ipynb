```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Sales Data - New - sales_data_sample.csv')
df
```

```python
df.info()
```

```python
# Remove duplicates and nulls 

cleaned_df = df.drop_duplicates(inplace = False)
cleaned_df
```

```python
# Add sales column

cleaned_df['SALES'] = cleaned_df['QUANTITYORDERED']*cleaned_df['PRICEEACH']
cleaned_df
```

```python
# Sales based on product line

productline_sales = cleaned_df.groupby(['PRODUCTLINE'], as_index = False)['SALES'].sum()
productline_sales.sort_values(by=['SALES'], inplace = True, ascending = False)
productline_sales.reset_index(drop = True, inplace = True)
display(productline_sales)

# Sales based on product line plot

fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(x = productline_sales['PRODUCTLINE'], y = productline_sales['SALES'], palette = sns.color_palette("BuGn_r", len(productline_sales['SALES'])))
```

```python
# Quantity based on product line

quantity_productline = cleaned_df.groupby(['PRODUCTLINE'], as_index = False)['QUANTITYORDERED'].sum()
quantity_productline.sort_values(by=['QUANTITYORDERED'], inplace = True, ascending = False)
quantity_productline.reset_index(drop = True, inplace = True)
display(quantity_productline)

# Quantity based on product line plot

fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(x = quantity_productline['PRODUCTLINE'], y = quantity_productline['QUANTITYORDERED'], palette = sns.color_palette("BuGn_r", len(quantity_productline['QUANTITYORDERED'])))
```

```python
# Average price based on product line

price_productline = cleaned_df.groupby(['PRODUCTLINE'], as_index = False)['PRICEEACH'].mean()
price_productline.sort_values(by=['PRICEEACH'], inplace = True, ascending = False)
price_productline.reset_index(drop = True, inplace = True)
display(price_productline)
```

```python
# Rearrange the orderdate to "year-month"

years = pd.to_datetime(cleaned_df['ORDERDATE'])
years = years.dt.to_period('M')
cleaned_df.loc[:, 'ORDERDATE'] = years
cleaned_df
```

```python
# Sales progress over a certain time period and the plot

sum_df = cleaned_df.groupby(['ORDERDATE'], as_index = False)['SALES'].sum()

fig, ax = plt.subplots(figsize=(30, 5))
sns.lineplot(x = sum_df['ORDERDATE'].astype(str), y = sum_df['SALES'], ax = ax, color = 'black')
plt.grid()
```

```python
# Each of productline sales progress over a certain time period and the plot
sum_productline_df = cleaned_df.groupby(['ORDERDATE', 'PRODUCTLINE'], as_index = False)['SALES'].sum()
sum_productline_df.sort_values(by=['PRODUCTLINE', 'ORDERDATE'], inplace = True)
sum_productline_df.reset_index(drop = True, inplace = True)

fig, ax = plt.subplots(figsize=(30, 5))
lab = []
sns.lineplot(x = sum_productline_df['ORDERDATE'].astype(str), 
                 y = sum_productline_df['SALES'], 
                 hue = sum_productline_df['PRODUCTLINE'], 
                 ax = ax)
plt.grid()
```

```python
# Each of deal size contribution to the total sales

fig, ax = plt.subplots(figsize=(10, 5))
dealsize_df = cleaned_df.groupby(['DEALSIZE'])['SALES'].sum()
value = []
for _, y in dealsize_df.items():
    value.append(y)
max_val = max(value)
min_val = min(value)
cols = []
for i in value:
    if i == max_val:
        cols.append('#2ca02c')
    elif i == min_val:
        cols.append('#ff7f0e')
    else:
        cols.append('#1f77b4')
plt.pie(dealsize_df, labels=list(dealsize_df.keys()), autopct='%.1f%%', colors = cols)
for i in ax.containers:
    ax.bar_label(i,)
```

```python
# Range of sales and average sales in each deal size

dealsize_sales_mean_df = cleaned_df.groupby(['DEALSIZE'], as_index = False)['SALES'].aggregate(['min','max', 'mean'])

display(dealsize_sales_mean_df)
```
